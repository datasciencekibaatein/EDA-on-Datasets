<h3>Clean and preparing the data for model training</h3>

```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv("Data/train.csv")
```

```python
df_test = pd.read_csv("Data/test.csv")
```

```python
df_train.info()
```

Append train and test data

```python
df = pd.concat([df_train,df_test],ignore_index=True)
df
```

```python
df.head()
```

```python
df.info()
```

drop user_id column its of no use

```python
df.drop(["User_ID"],axis=1,inplace=True)
```

Convert Gender into zero and one, F = 0 and M = 1

```python
df["Gender"] = df["Gender"].map({'F':0,"M":1})
```

Handle Age column, it's categorical feature

```python
df["Age"].unique()
```

```python
#ordinal encoding using map function
df["Age"] = df["Age"].map({'0-17':1, '55+':7, '26-35':3, '46-50':5, '51-55':6, '36-45':4, '18-25':2})
```

Handling city category

```python
df["City_Category"].unique()
```

```python
df_city = pd.get_dummies(df["City_Category"])
```

```python
df =pd.concat([df,df_city],axis=1)
```

```python
df.drop(["City_Category"],inplace=True,axis=1)
```
