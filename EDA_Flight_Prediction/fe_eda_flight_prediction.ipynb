<h1>EDA and Feature Engineering on Flight prediction dataset</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("flight-price.xlsx",)
```

```python
## Feature engineering
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
## Convert date,month and year in numerical value

df['Date'] = df['Date'].astype(int)
df['Month']= df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
#drop date_of_journey column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
#split arrival time

df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
## Convert into integer

df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
## handling departure time
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_min'] = df['Dep_Time'].str.split(':').str[1]
df['Departure_hour'] = df['Departure_hour'].astype(int)
df['Departure_min'] = df['Departure_min'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
#Total stops column , dealing with categorical data

df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.nan:1, '4 stops':4})
```

```python
df.head()
```

```python
#Fixing Duration
df['Duration_hour'] = df['Duration'].str.split(" ").str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(" ").str[1].str.split('m').str[0]
```

```python
df['Duration_hour'] = df['Duration_hour'].astype(int)
df['Duration_min'] = df['Duration_min'].astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
#drop route
df.drop('Route',axis=1,inplace=True)
```

```python
#encoding 
from sklearn.preprocessing import OneHotEncoder
```

```python
df[['Airline', 'Source', 'Destination', 'Additional_Info']]
```

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

encoded=encoder.fit_transform( df[['Airline', 'Source', 'Destination', 'Additional_Info']])
```

```python
encoded
```

```python
encoded_col_names = encoder.get_feature_names_out(
    ['Airline', 'Source', 'Destination', 'Additional_Info']
)
```

```python
encoded_df = pd.DataFrame(encoded,columns=encoded_col_names)
encoded_df
```

```python
## drop source, destination,airline, and additional_infor
df.drop('Airline',axis=1,inplace=True)
df.drop('Source',axis=1,inplace=True)
df.drop('Destination',axis=1,inplace=True)
df.drop('Additional_Info',axis=1,inplace=True)

```

```python
df.head()
```

```python
final_df = pd.concat([encoded_df,df])
```
